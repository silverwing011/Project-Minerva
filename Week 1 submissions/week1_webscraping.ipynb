```python
import pandas as pd
import requests
import json
from bs4 import BeautifulSoup
```

# adding news website link

```python
website_link = "https://www.indiatvnews.com/"
browser_header = {
    "User-Agent": "Mozilla/5.0"
}
page = requests.get(
    website_link,
    headers=browser_header
)
print(page.status_code)
```

# Parsing html page

```python
soup = BeautifulSoup(
    page.text,
    "html.parser"
)
print(soup.title.text)
```

# Collecting headlines

```python
headline_tags = soup.find_all("h3")
collected_headlines = []

for item in headline_tags:
    headline_text = item.get_text(strip=True)

    if headline_text:
        collected_headlines.append(headline_text)
print(collected_headlines[:10])
```

```python
with open("raw_news_data.txt","w",encoding="utf-8"
) as file:
    for headline in collected_headlines:
        file.write(headline + "\n")
print("text file saved")
```

```python
with open("news_headlines.json","w",encoding="utf-8"
) as file:
    json.dump(
        collected_headlines,
        file,
        indent=4
    )
print("json file saved")
```

```python
headline_dataframe = pd.DataFrame(
    collected_headlines,
    columns=["News Headlines"]
)
headline_dataframe.head()
```

```python
headline_dataframe.to_csv(
    "headline_dataset.csv",
    index=False
)
print("csv file saved")
```

# tokenizing headlines

```python
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')
```

```python
nltk.download('punkt_tab')
```

```python
sample_text = " ".join(collected_headlines)
tokens = word_tokenize(sample_text)
print(tokens[:50])
```

```python
print(len(tokens))
```

# sentence tokenization

```python
from nltk.tokenize import sent_tokenize
sentences = sent_tokenize(sample_text)
print(sentences[:5])
```

# embeddings

```python
from sentence_transformers import SentenceTransformer
```

```python
model = SentenceTransformer(
    'all-MiniLM-L6-v2'
)
```

```python
embeddings = model.encode(sentences)
print(embeddings.shape)
```

```python
import numpy as np
np.save(
    "headline_embeddings.npy",
    embeddings
)
print("embeddings saved")
```
