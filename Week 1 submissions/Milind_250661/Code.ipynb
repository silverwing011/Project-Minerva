```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://timesofindia.indiatimes.com/sports/cricket/news/virat-kohli-test-comeback-loading-fans-erupt-after-childhood-coachs-cryptic-lets-see-what-happens-remark/articleshow/131243570.cms"
```

```python
response = requests.get(url)
response.status_code
```

```python
soup=BeautifulSoup(response.content, 'html.parser')
soup.title.text
```

```python
soup.text
```

```python
len(soup.text)
```

```python
paragraphs = soup.find_all('p')
len(paragraphs)
```

```python
article_text = ' '
for p in paragraphs:
    text=p.get_text(strip=True)
    article_text += text + ' '
article_text[:2000]
```

```python
len(article_text)
```

```python
with open("news_corpus.txt", "w", encoding="utf-8") as f:
    f.write(article_text)
```

```python
df = pd.DataFrame({ 
    "url": [url],
    "text": [article_text]
})

df.to_csv("news_corpus.csv", index=False)

df.head()
```

```python
import json

data = { "url": url, "title": soup.title.text, "text": article_text}

with open("news_corpus.json", "w", encoding="utf-8") as f:
    json.dump(data, f, indent=4, ensure_ascii=False)
```
