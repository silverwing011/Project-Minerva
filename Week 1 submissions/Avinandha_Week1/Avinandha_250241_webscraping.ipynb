Implementation task

Step 1 : Fetching the html content

```python
import requests

url = "https://www.hindustantimes.com/india-news/i-too-am-a-cockroach-oppn-leaders-endorse-cockroach-janata-party-say-it-s-bjp-vs-cjp-101779343530465.html"
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64)'}
response = requests.get(url, headers=headers)
```

Step 2: Parsing the HTML

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(response.text, "html.parser")

contents = soup.find_all('p', class_ = 'content')
```

Extracting the text content and creating ther csv file

```python
import csv

filename = "corpusX.csv"

with open(filename, mode='w', newline='', encoding='utf-8') as file:
    writer = csv.DictWriter(file, fieldnames=['Article'])

    writer.writeheader()
    for content in contents:
      text = content.get_text()
      writer.writerow({'Article': text})
```

Implementation task 2 : tokenising and embeddding

```python
import pandas as pd

data = pd.read_csv('corpusX.csv')
```

```python
import nltk
nltk.download('punkt_tab')
from nltk.tokenize import sent_tokenize
corpus = ""
for line in data['Article']:
  sentences = sent_tokenize(line)
  corpus += " ".join(sentences) + " "

print(corpus)
```

```python
all_sentences = sent_tokenize(corpus)

print(all_sentences)
```

Embeding

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer("sentence-transformers/all-MiniLM-L6-v2")
embeddings = model.encode(all_sentences)
print(embeddings.shape)
```

```python
similarities = model.similarity(embeddings, embeddings)
print(similarities)
```
